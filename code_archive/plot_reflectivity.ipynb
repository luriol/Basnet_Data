```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import os 


data_dir = Path("../data_archive")
result_dir = Path("../result_archive")
plot_dir = Path("../figure_archive")
os.listdir(data_dir)

# Load the saved plot data
df_plot = pd.read_csv(data_dir / "reflectivity_data_neat.csv")
# Plot "neat" data
plt.errorbar(df_plot["q_nm"], df_plot["Ic_neat"], df_plot["dIc_neat"], fmt='ko', markersize=6, 
             markerfacecolor='none', markeredgecolor='k', label="Neat Data")
plt.plot(df_plot["q_nm"], df_plot["yf_neat"], '--r', label="Neat Fit")

df_plot = pd.read_csv(data_dir / "reflectivity_data_gold.csv")
# Plot "gold" data
plt.errorbar(df_plot["q_nm"], df_plot["Ic_gold"], df_plot["dIc_gold"], fmt='k^', markersize=6, 
             markerfacecolor='none', markeredgecolor='k', label="Gold Data")
plt.plot(df_plot["q_nm"], df_plot["yf_gold"], '-.r', label="Gold Fit")

# Set labels, log scale, and limits
plt.xlabel('Q (nm$^{-1}$)', fontsize=14, fontweight='bold')
plt.ylabel('Intensity (cps)', fontsize=14, fontweight='bold')
plt.tick_params(axis='x', labelsize=12)
plt.tick_params(axis='y', labelsize=12)
plt.yscale('log')
plt.xlim(0, 5)
plt.ylim(2e-9, 1.2)
# plt.legend()

# Save and show plot
plt.savefig(plot_dir / 'reflectivity.jpg')
plt.show()
```
